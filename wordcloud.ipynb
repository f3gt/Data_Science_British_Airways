```python
df=pd.read_csv(r'E:\learn\british_airways_text_review.csv')
```

```python
df
```

```python
stopword=set(STOPWORDS)
comment_word=''
stopword.update(["ba","flight", "british","airway", "airline","plane", "told","also","passenger" \
                 "london", "heathrow", "aircraft", "could","even", "would","used","row","bag","point","return","better","made",
                "one","1hr","will","still","given"])
```

```python
for val in df.Review:
    val=str(val)
    
    token=val.split()
    
    
    for i in range(len(token)):
        token[i]=token[i].lower()
    comment_word += " ".join(token)+" "
    
    

```

```python
wordcloud=WordCloud(font_path=None,
    width=800,
    height=800,
    background_color='white',
    ranks_only=None,
    stopwords=stopword,
    max_words=500).generate(comment_word)

plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
